## 1. My rule and its reason codes

*Write the rule in plain words first. Then the reason codes it can output.*

A product listing on M.Minhas Watches is worth reviewing if it gets decent traffic has not been updated recently and its sales rank is slipping.Reason Codes: stale_but_visible,Position_slipping.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
data={'Product_id':[101,102,103],'days_since_update':[200,50,190],'impressions':[600,200,800]}
df=pd.DataFrame(data)
stale=(df['days_since_update']>=180).astype(int)
visible=(df['impressions']>=500).astype(int)
df['score']=stale*visible*df['impressions']
print("Baseline Rule and scores Generated:\n",df)
print("Reason codes attached :['stale_but_visible','position_slipping']")
```

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

Ranked the inventory items based on the transparent baseline score and exported the sorted queue to"outputs/baseline_action_score.csv"

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import os
import pandas as pd
os.makedirs('outputs',exist_ok=True)
ranked_df=df.sort_values(by='score',ascending=False)
output_path='outputs/baseline_action_score.csv'
ranked_df.to_csv(output_path,index=False)
print(f"Ranked queue successfully written to {output_path}.\n")
print(ranked_df)
```
